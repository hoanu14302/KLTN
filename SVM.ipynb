```python
import pandas as pd

data_covid = pd.read_csv("D:/KLTN/dataset_covid.csv", delimiter=",", encoding="latin-1")
data_covid
```

TẠO DATAFRAME

- X lấy từ cột thứ 3 (0 - 2) và trừ cột cuối cùng == feature
- y lấy cột cuối cùng làm nhãn == label


```python
# Sử dụng iloc để lấy toàn bộ cột và dòng
X = data_covid.iloc[:, 3:-1]  # Lấy từ cột thứ 3 đến cột trước cột cuối cùng
y = data_covid.iloc[:, -1]  # Lấy cột cuối cùng làm nhãn

df = pd.concat([X, y], axis=1)  # Tạo DataFrame từ X và y

display(df)  # Hiển thị DataFrame
```

```python
X_df_processed = df_processed.iloc[:, :-1]
y_df_processed = df_processed.iloc[:, -1]

X_df_processed.info()
y_df_processed.info()
```

MÔ HÌNH KNN dạng phân lớp


SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd

C_values = [10.0, 100.0]
kernel_values = ['linear', 'rbf', 'sigmoid']
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

X = X_df_processed.to_numpy()
y = y_df_processed.to_numpy().ravel()

evaluation_values = {
    'C': [],
    'Kernel': [],
    'Mean Accuracy': [],
    'Mean Precision': [],
    'Mean Recall': [],
    'Mean F1 Score': []
}

for C in C_values:
    for kernel in kernel_values:
        accuracy_scores = []
        precision_scores = []
        recall_scores = []
        f1_scores = []

        for train_index, test_index in skf.split(X, y):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]

            svm = SVC(C=C, kernel=kernel, probability=True)
            svm.fit(X_train, y_train)
            y_pred_proba = svm.predict_proba(X_test)

            y_pred = np.argmax(y_pred_proba, axis=1)

            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='micro')
            recall = recall_score(y_test, y_pred, average='micro')
            f1 = f1_score(y_test, y_pred, average='micro')

            accuracy_scores.append(accuracy)
            precision_scores.append(precision)
            recall_scores.append(recall)
            f1_scores.append(f1)

        mean_accuracy = np.mean(accuracy_scores)
        mean_precision = np.mean(precision_scores)
        mean_recall = np.mean(recall_scores)
        mean_f1_score = np.mean(f1_scores)

        print("C:", C)
        print("Kernel:", kernel)
        print("Mean Accuracy:", mean_accuracy)
        print("Trung bình Accuracy:", mean_accuracy)
        print("Trung bình Precision:", mean_precision)
        print("Trung bình Recall:", mean_recall)
        print("Trung bình F1 Score:", mean_f1_score)
        print("------------------------------")
        
        evaluation_values['C'].append(C)
        evaluation_values['Kernel'].append(kernel)
        evaluation_values['Mean Accuracy'].append(mean_accuracy)
        evaluation_values['Mean Precision'].append(mean_precision)
        evaluation_values['Mean Recall'].append(mean_recall)
        evaluation_values['Mean F1 Score'].append(mean_f1_score)

# Tạo DataFrame từ các giá trị đánh giá
df = pd.DataFrame(evaluation_values)

# Lưu DataFrame thành file CSV
df.to_csv('D:\KLTN\Danh_gia_mo_hinh\ket_qua_danh_gia_SVM_2.csv', index=False)
```
