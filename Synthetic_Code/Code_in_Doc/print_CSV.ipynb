```python
import matplotlib.pyplot as plt
import pandas as pd

# Đọc dữ liệu từ tệp "gdp_per_capita.csv"
gdp_per_capita = pd.read_csv("gdp_per_capita.csv", thousands=',', delimiter=',',
                             encoding='latin1', na_values="n/a")

# Lấy các cột "Country" và "2015" từ DataFrame
countries = gdp_per_capita["Country"]
gdp_2015 = gdp_per_capita["2015"]

# Tạo một hình vẽ mới với kích thước 10x6 inches
plt.figure(figsize=(50, 50))

# Vẽ đồ thị
plt.bar(countries, gdp_2015)
plt.xlabel("Country")
plt.ylabel("GDP per capita (2015)")
plt.xticks(rotation=90)
plt.show()
```
