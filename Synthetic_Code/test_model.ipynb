```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Đọc dữ liệu từ tệp tin CSV vào DataFrame
data = pd.read_csv("D:/KLTN/gdp_per_capita.csv", encoding="latin-1", delimiter=",")
```

```python
data
```

```python
# Lấy cột chứa dữ liệu văn bản
text_data = data["Country"].values
```

```python
# Khởi tạo một đối tượng CountVectorizer để vector hóa dữ liệu
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(text_data)

# In ra danh sách các đặc trưng
feature_names = vectorizer.get_feature_names_out()
feature_names
```
