KNN

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd

k_values = [1, 3, 5, 7, 9]
n_splits = 10
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42) #Tạo ra 10 fold và xáo trộn dữ liệu (shuffle) 

# Chuyển dữ liệu trong df thành dạng numpy
X = X_df_processed.to_numpy()
y = y_df_processed.to_numpy().ravel()

# Lưu các kết quả để tạo thành file csv
evaluation_values = {
    'K': [],
    'Mean Accuracy': [],
    'Mean Precision': [],
    'Mean Recall': [],
    'Mean F1 Score': []
}

for k in k_values:
    # Tạo array để lưu các kết quả --> Tính trung bình các đánh giá
    accuracy_scores = [] 
    precision_scores = [] 
    recall_scores = []
    f1_scores = []

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Tạo mô hình KNN
        knn = KNeighborsClassifier(n_neighbors=k, metric='minkowski')
        knn.fit(X_train, y_train)            
        y_pred_proba = knn.predict_proba(X_test) # dự đoán xác suất cho các lớp đích của các mẫu dữ liệu kiểm tra X_test

        y_pred = np.argmax(y_pred_proba, axis=1) # chọn lớp có xác suất cao nhất cho mỗi mẫu dữ liệu

        accuracy = accuracy_score(y_test, y_pred) # tỷ lệ các mẫu dữ liệu được phân loại đúng (nhãn dự đoán khớp với nhãn thực tế) so với tổng số mẫu dữ liệu
        precision = precision_score(y_test, y_pred, average='micro') # tính toán precision trên toàn bộ dữ liệu
        recall = recall_score(y_test, y_pred, average='micro')
        f1 = f1_score(y_test, y_pred, average='micro')
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)

    mean_accuracy = np.mean(accuracy_scores)
    mean_precision = np.mean(precision_scores)
    mean_recall = np.mean(recall_scores)
    mean_f1_score = np.mean(f1_scores)

    print("K:", k)
    print("Trung bình Accuracy:", mean_accuracy)
    print("Trung bình Precision:", mean_precision)
    print("Trung bình Recall:", mean_recall)
    print("Trung bình F1 Score:", mean_f1_score)
    print("------------------------------")

    evaluation_values['K'].append(k)
    evaluation_values['Mean Accuracy'].append(mean_accuracy)
    evaluation_values['Mean Precision'].append(mean_precision)
    evaluation_values['Mean Recall'].append(mean_recall)
    evaluation_values['Mean F1 Score'].append(mean_f1_score)

# Tạo DataFrame từ các giá trị đánh giá
df = pd.DataFrame(evaluation_values)

# Lưu DataFrame thành file CSV
df.to_csv('D:\KLTN\Danh_gia_mo_hinh\ket_qua_danh_gia_KNN.csv', index=False)
```

SVM

```python
from sklearn.svm import SVC

C_values = [0.1, 1.0]
kernel_values = ['linear', 'rbf', 'sigmoid']
n_splits = 10
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

X = X_df_processed.to_numpy()
y = y_df_processed.to_numpy().ravel()

evaluation_values = {
    'C': [],
    'Kernel': [],
    'Mean Accuracy': [],
    'Mean Precision': [],
    'Mean Recall': [],
    'Mean F1 Score': []
}

for C in C_values:
    for kernel in kernel_values:
        accuracy_scores = []
        precision_scores = []
        recall_scores = []
        f1_scores = []

        for train_index, test_index in skf.split(X, y):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]

            svm = SVC(C=C, kernel=kernel, probability=True)
            svm.fit(X_train, y_train)
            y_pred_proba = svm.predict_proba(X_test)

            y_pred = np.argmax(y_pred_proba, axis=1)

            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='micro')
            recall = recall_score(y_test, y_pred, average='micro')
            f1 = f1_score(y_test, y_pred, average='micro')

            accuracy_scores.append(accuracy)
            precision_scores.append(precision)
            recall_scores.append(recall)
            f1_scores.append(f1)

        mean_accuracy = np.mean(accuracy_scores)
        mean_precision = np.mean(precision_scores)
        mean_recall = np.mean(recall_scores)
        mean_f1_score = np.mean(f1_scores)

        print("C:", C)
        print("Kernel:", kernel)
        print("Mean Accuracy:", mean_accuracy)
        print("Trung bình Accuracy:", mean_accuracy)
        print("Trung bình Precision:", mean_precision)
        print("Trung bình Recall:", mean_recall)
        print("Trung bình F1 Score:", mean_f1_score)
        print("------------------------------")
        
        evaluation_values['C'].append(C)
        evaluation_values['Kernel'].append(kernel)
        evaluation_values['Mean Accuracy'].append(mean_accuracy)
        evaluation_values['Mean Precision'].append(mean_precision)
        evaluation_values['Mean Recall'].append(mean_recall)
        evaluation_values['Mean F1 Score'].append(mean_f1_score)

# Tạo DataFrame từ các giá trị đánh giá
df = pd.DataFrame(evaluation_values)

# Lưu DataFrame thành file CSV
df.to_csv('D:\KLTN\Danh_gia_mo_hinh\ket_qua_danh_gia_SVM_1.csv', index=False)
```
